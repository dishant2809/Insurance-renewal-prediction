```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
train = train.drop("id",1)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train.describe()
```

```python
train['Count_3-6_months_late'].fillna(train['Count_3-6_months_late'].std(),inplace=True)
train['Count_6-12_months_late'].fillna(train['Count_6-12_months_late'].std(),inplace=True)
train['Count_more_than_12_months_late'].fillna(train['Count_more_than_12_months_late'].std(),inplace=True)
train['application_underwriting_score'].fillna(99.21,inplace=True)
```

```python
test.isnull().sum()
```

```python
test.describe()
```

```python
test['Count_3-6_months_late'].fillna(test['Count_3-6_months_late'].std(),inplace=True)
test['Count_6-12_months_late'].fillna(test['Count_6-12_months_late'].std(),inplace=True)
test['Count_more_than_12_months_late'].fillna(test['Count_more_than_12_months_late'].std(),inplace=True)
test['application_underwriting_score'].fillna(98.8,inplace=True)
```

```python
test.info()
```

```python
train = pd.get_dummies(train)
```

```python
test = pd.get_dummies(test)
```

```python
train.head()
```

```python
test.head()
```

```python
X = train.drop("target",1)
```

```python
y = train['target']
```

```python
y.head()
```

```python
test2 = test.drop('id',1)
```

```python
test2.head()
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lreg.fit(X,y)
pred_a = lreg.predict(test2)
```

```python
submission = pd.DataFrame({
        "id": test["id"],
        "target": pred_a
    })

submission.to_csv('4ilreg.csv', index=False)
```
